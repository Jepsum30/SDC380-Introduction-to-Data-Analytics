```python
# Code Cell 1: Import libraries and load the dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the Excel file (ensure the file is in your working directory)
df = pd.read_excel('Happiness_Development_Index (2).xlsx', sheet_name='HDI')

# Display first few rows to verify
print(df.head())
print(df.columns)  # Check column names
```

```python
# Code Cell 2: Histogram of HDI scores
plt.figure(figsize=(10, 6))
plt.hist(df['Human Development Index (HDI) '], bins=25, edgecolor='black', alpha=0.75, color='skyblue')
plt.title('Distribution of HDI Scores Across Countries')
plt.xlabel('Human Development Index (HDI)')
plt.ylabel('Number of Countries')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# Code Cell 3: Bar chart of top 10 countries by HDI
# Sort by HDI descending and take top 10
top10 = df.nlargest(10, 'Human Development Index (HDI) ')

plt.figure(figsize=(12, 8))
plt.barh(top10['Country'], top10['Human Development Index (HDI) '], color='teal')
plt.title('Top 10 Countries with Highest HDI Scores')
plt.xlabel('HDI Score')
plt.ylabel('Country')
plt.gca().invert_yaxis()  # Highest on top
plt.grid(axis='x', alpha=0.3)
plt.show()
```

```python
# Code Cell 4: Scatter plot of HDI vs GNI per capita
plt.figure(figsize=(10, 6))
plt.scatter(df['Gross national income (GNI) per capita'], 
            df['Human Development Index (HDI) '], 
            alpha=0.7, color='coral', edgecolor='black')
plt.title('Relationship Between HDI and GNI per Capita')
plt.xlabel('Gross National Income (GNI) per Capita')
plt.ylabel('Human Development Index (HDI)')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Code Cell 5: Heatmap of correlations with HDI and other numerical variables
# Select numerical columns (excluding Rank if it's not needed)
numerical_df = df.select_dtypes(include=[np.number])

# Compute correlation matrix
corr_matrix = numerical_df.corr()

# Create heatmap focused on correlations
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='RdBu_r', vmin=-1, vmax=1, 
            center=0, fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of HDI with Other Numerical Variables')
plt.show()
```
